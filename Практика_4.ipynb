```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/EvgeniiSopovGitHub/datasets_uni/main/okaivd/db_students_performance_delim.csv",sep=';',decimal = ',')
print(df.head())
print(df.info())
print(df.tail())

df = df.fillna(df.mean())
df['min'] = df.iloc[:, :10].min(axis=1)
df['median'] = df.iloc[:, :10].median(axis=1)
df['max'] = df.iloc[:, :10].max(axis=1)
mean_value = round(df.iloc[:, :10].mean(axis=1), 2)
std_value = round(df.iloc[:, :10].std(axis=1), 2)
df['mean/std'] = mean_value.astype(str) + '/' + std_value.astype(str)
df.loc['average annual'] = df.iloc[:, :10].mean()

df.to_csv("lab4_bsvv 22 - 01_danilov.csv")
df.head()
```
